### Drop the first row and NaN

```
print(type(df[0]))
df = df[0]
df.rename({0: 'Postcode', 1: 'Borough', 2: 'Neighbourhood'}, axis = 'columns', inplace = True)
```

```
print(df.isnull().sum())
df.drop(df.index[0], inplace = True)
df.dropna(axis = 0, inplace = True)
print(df.isnull().sum())
df
```

```
print('The dataframe has {} boroughs and {} neighbourhoods.'.format(

len(df['Borough'].unique()),

len(df['Neighbourhood'].unique())))
print(df.shape)
df.to_csv('Toronto_df.csv')
```

```
df = pd.read_csv('Toronto_df.csv')
```

###Acquire the postal codes: 

```
df['Postcode'].unique()
```

```
#sort the dataframe and drop its index
df = df.sort_values(by = ['Postcode'], ascending = True)
df.reset_index(inplace= True, drop = True)
df.head()
```

### Use Geocoder package to get the geographical coordinates of the neighborhoods

```
pip install geocoder
```

```
import geocoder

latitude = []
longitude = []
n=211
postal_code = df['Postcode']

for j in range(n):
    # initialize your variable to None
    lat_lng_coords = None
    
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.google('{}, Toronto, Ontario'.format(postal_code[j]))
        lat_lng_coords = g.latlng

    latitude.append(lat_lng_coords[0])
    longitude.append(lat_lng_coords[1])
    
df['Latitude'], df['Longitude'] = latitude, longitude

df
```
